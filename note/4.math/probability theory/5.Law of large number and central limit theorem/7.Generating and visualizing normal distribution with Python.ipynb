```python
from scipy.stats import norm
# loca is expected value, scale is standard deviation
X = norm(loc = 1, scale=2)
```

```python
# generate a sample of size 10
X.rvs(10)
```

```python
X.pdf(1.2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(-4,6, 200)
plt.plot(x, X.pdf(x))
smaple = X.rvs(1000)
plt.hist(smaple, density=True, bins=30)
```

```python
plt.plot(x, norm.pdf(x, loc=0,scale=1))
plt.plot(x, norm.pdf(x, loc=1,scale=1))
plt.plot(x, norm.pdf(x, loc=1,scale=2))
```

```python
# sample_size=1
sample_size=5
# sample_size=10
# sample_size=100
number_of_sample=10000
means = []
zscores = []
EX = 0.5
VarX = 1./12

for i in range(number_of_sample):
    smaple = np.random.uniform(low=0, high=1, size=sample_size)
    
    mean = sum(smaple) / sample_size
    zscore = (mean-EX) * np.sqrt(sample_size) / np.sqrt(VarX)
    means.append(mean)
    zscores.append(zscore)
```

```python
# xlim is limit of x axis
plt.xlim(-4, 4)
plt.hist(zscores, density=True, bins=50);

x = np.linspace(-4, 4, 1000)

plt.plot(x, norm.pdf(x))
```
