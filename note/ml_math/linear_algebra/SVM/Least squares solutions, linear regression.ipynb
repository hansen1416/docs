Let $Ax = b$ be an inconsistent system of linear equations, where $A$ is $m \times n$

$x=\begin{pmatrix} x_1 \\ \vdots \\ x_n \end{pmatrix} \in \mathbb{R}^n,
b=\begin{pmatrix} b_1 \\ \vdots \\ b_m \end{pmatrix} \in \mathbb{R}^m$

What is the best approximation $\hat{x}$ for a solution?

Given $x \in \mathbb{R}^n$, let $D(x)$ denotes

$ |Ax-b|^2 =(\langle a_1,x \rangle - b_1)^2 + \dotsm + (\langle a_m, x \rangle - b_m)^2 $

Then let $\hat{x}$ be such that $D(\hat{x}) = min$

Let $L = R(A)$ be the subspace of $\mathbb{R}^m$ generated by the columns $A^1,\dots,A^n$ of $A$. Then $Ax \in L$ and $D(x) = d(b, Ax)^2$

Hence $D(x)=min \;if\; d(b, Ax) = d(b,L)$

This happens if $Ax = pr_Lb$. Then $D(x)=|ort_Lb|^2$

_Proposition_ Suppose that $A$ is and $m \times n$ matrix with rank $A=n$, and $b \in \mathbb{R}^m$. The the least sqaure solution of the system $Ax=b$ is

$\hat{x} = pr_L b = (A^T A)^{-1}A^Tb$

Then $A\hat{x}=pr_L b$ and $D(\hat{x})=|ort_Lb|^2$

_Proof_. We have: $a=\{A^1,\dots,A^n\}$ is a basis of $L$

Then $D(x) = min = |ort_Lb|^2$ only if $Ax=pr_Lb$

Note: $\hat{x} = (pr_Lb)_a$ is unique such that $A\hat{x} = pr_L b$

For $x=(A^T A)^{-1}A^Tb$, we have

$Ax=A(A^TA)^{-1}A^Tb = pr_Lb$

So, $\hat{x} = (A^TA)^{-1}A^Tb$ (We are trying to minimize the distance between Ax and b)

Then $A\hat{x}=pr_Lb$ and

$D(\hat{x})=|b-A\hat{x}|^2 = |b-pr_Lb|^2=|ort_Lb|^2$

__Linear regression__

Given $x^1,\dots,x^n \in \mathbb{R}^m$ and $y_1,\dots,y_n$, we would like to find $c \in \mathbb{R}^m$ that minimise the error of "equalities"

$\langle x^i, c\rangle \approx y_i, i = 1,\dots,n $

Let $A = (x^1, \dots , x^n)^T \in \mathbb{R}^{n \times m}$. We put $c = \hat{c}$, the least square solution of $Ac=y$

So $\hat{c} = (A^TA)^{-1}A^Ty$

For $m=1$, we have $A=(x^1, \dots, x^n)^T$ and $\hat{c} = \langle A, A\rangle^{-1} A^T y = \frac{\langle A, y\rangle}{\langle A,A \rangle} = \frac{\langle A, y\rangle}{|A|^2}$
