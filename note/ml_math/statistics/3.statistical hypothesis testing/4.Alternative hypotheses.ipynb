$X \sim Bin(n,p) $

$H_0 : p = \frac{1}{2}$

$H_1 : p \ne \frac{1}{2}$

$X_{obs}$ is a sampled value from $X$

$|x_{critical}|$

$P(|X-\frac{n}{2}| > x_{critical} \le \alpha$

$\implies P(X > x_{critical}) \le \frac{\alpha}{2}$

reject $H_0$ if $|X_{obs} - \frac{n}{2}| > x_{critical}$

p-value$(X_{obs}) = P(|X-\frac{n}{2}|) > |X_{obs} - \frac{n}{2}|$

p-value is the area defined by $-\infty, -X_{obs}, X_{obs}, +\infty$

_example_

The magician claims he can predict future but at some days his vision of future is inversed, i.e. if the coin will land with head he would predict tail and vise versa. He cannot detect in advance wether today is an "ordinary" or "inverse" day. Therefore, he claims, very small number of guessed results is also an evidence in favor for his magical abilities. It means that now an alternative is not $p>1/2$ but $p\ne 1/2$. Now we update our testing procedure to take it into account. Assume that we toss a coin 11 times (during the same day). Let us pick such value $k$ that if the magician guess $k$ times or less, or $(11-k)$ times or more, we reject null hypothesis. For example, if $k=0$, the magician has to guess exactly 11 times or exactly 0 times to prove his abilities. Find the largest value $k$ such that probability to reject null hypothesis provided that null hypothesis holds is less than the significance level 0.05.


$P(X=11 | H_0) = P(X=0 | H_0) = \binom{11}{11} \times \frac{1}{2}^{11} \times \frac{1}{2}^0 = 1 \times \frac{1}{2048} = \frac{1}{2048}$

$P(k=0) = \frac{1}{2048} + \frac{1}{2048} = \frac{2}{2048}$

$P(X=10 | H_0) = P(X=1 | H_0) = \binom{11}{10} \times \frac{1}{2}^{10} \times \frac{1}{2} = 11 \times \frac{1}{2048} = \frac{11}{2048}$

$P(k=1) = \frac{1}{2048} + \frac{1}{2048} + \frac{11}{2048} + \frac{11}{2048} = \frac{24}{2048} = 0.01171875$

$P(X=9 | H_0) = P(X=2 | H_0) = \binom{11}{9} \times \frac{1}{2}^9 \times \frac{1}{2}^2 = 55 \times \frac{1}{512} \times \frac{1}{4} = \frac{55}{2048}$

$P(k=2) = \frac{1}{2048} + \frac{1}{2048} + \frac{11}{2048} + \frac{11}{2048} + \frac{55}{2048} + \frac{55}{2048} = \frac{134}{2048} = 0.065429688 > 5\%$

k = 1

_example_

Assume that we perform the following test. $X$ is binomially distributed random variable with probability of success pp and number of trials nn and $X_{obs}$ is an observed value of this random variable. Null hypothesis is that $p=1/2$. Consider two possible alternatives: $H_1: p>1/2$ and $H_1': p\ne 1/2$. If we test an alternative $H_1$ (one-sided), we get p-value $0.03$. Is it true that $X_{obs} > n/2$?

Yes

_If p-value is small and alternative is one-sided, it means that value of $X_{obs}$ is close to the corresponding tail of the distribution._
