If we will increase n further this shape will be more and more narrow, and it will be more and more concentrated near the expected value. __Z-score__

Z-score is in a sense, transformed version of this average. $\overline{X}_n = \frac{X_1 + \dots + X_n}{n}$

$Z_n = {\large \frac{\overline{X}_n-\mathbb{E}X}{\sqrt{Var X}} \sqrt{n} }$

$\overline{X}_n-\mathbb{E}X$ is shifting the random variable, let expected value =0

$\sqrt{Var X}$ and $\sqrt{n}$  
Then we multiply the result by some constant which depends on variance of x and n. This allows us to standardize  variance of this $Z_n$ in such a way that this variance equals to one. 

---

$\mathbb{E}Z_n = \mathbb{E}(\frac{\sqrt{n}}{\sqrt{Var X}} \cdot (\overline{X}_n - \mathbb{E}X)) 
= \frac{\sqrt{n}}{\sqrt{Var X}} \mathbb{E}(\overline{X}_n - \mathbb{E}X) 
= \frac{\sqrt{n}}{\sqrt{Var X}} (\mathbb{E}X - \mathbb{E}X) = 0$

$Var Z_n = Var (\frac{\sqrt{n}}{\sqrt{Var X}} \cdot (\overline{X}_n - \mathbb{E}X)) $
$= \frac{n}{Var X} \cdot Var (\overline{X}_n - \mathbb{E}X)$ shit doesn't change variance  
$= \frac{n}{Var X} \cdot Var (\overline{X}_n)$ in previous chapter $Var(\overline{X}_n) = \frac{Var X}{n}$  
$= \frac{n}{Var X} \cdot \frac{Var X}{n} = 1$

PDF of standard normal distribution

$Pdf_{N(0,1)}(x) = \frac{1}{\sqrt{2 \pi}} e^{-\frac{x^2}{2}}$

$N(0,1) \implies \mathbb{E} = 0, Var = 1$ 

So the central limit theorem says that cumulative distribution function of Z_n add some arbitrary point x tends to cumulative distribution function of standard normal distribution at the same point x. Here as usual, cumulative distribution function of this normal distribution can be calculated as integral of this probability density function.

__Central limit theorem__

$CDF_{Z_n}(x) \to CDF_{N(0,1)}(x) = \int^{x}_{-\infty} \frac{1}{2\pi} e^{-\frac{x^2}{2}}dx$
