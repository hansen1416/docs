>Chebyshev's inequality asks, what is the probability that the value of random variable is too far from its expected value?  
>Of course, this should be related to the variance, but Chebyshev's inequality gives us some explicit estimate.

$P(|X-\mathbb{E}X| > \alpha) \le {\large \frac{VarX}{\alpha^2} }, \alpha > 0$

_proof_

shift X to make $\mathbb{E}X=0$

Let $Y = X - \mathbb{E}X$, $\mathbb{E}Y=0$ and $Var Y=Var X$

$Var Y = \mathbb{E}(Y-0)^2 = \mathbb{E}(Y^2)$

Assume Y has PDF, $p(y)=PDF_Y(y)$

$\mathbb{E}(Y^2) = \int^{+\infty}_{-\infty} y^2 p(y)dy$



$\int^{+\infty}_{-\infty} y^2 p(y)dy \ge \int^{-\alpha}_{-\infty} \alpha^2 p(y)dy + \int^{\infty}_{\alpha} \alpha^2 p(y)dy
=\alpha^2(\int^{-\alpha}_{-\infty} p(y)dy + \int^{\infty}_{\alpha} p(y)dy)
$

The left integral equals to the probability that the value of y is less than negative Alpha and the right integral is a probability that value of y is larger than plus Alpha. So the sum of these probabilities gives us the probability that absolute value of y is larger than Alpha.

$\int^{-\alpha}_{-\infty} p(y)dy = P(y<-\alpha)$

$\int^{\infty}_{\alpha} p(y)dy = P(y>\alpha)$

$\alpha^2(\int^{-\alpha}_{-\infty} p(y)dy + \int^{\infty}_{\alpha} p(y)dy) = \alpha^2(P(|Y| > \alpha))$

finally

$Var Y \ge \alpha^2(P(|Y| > \alpha)) \implies P(|Y| > \alpha) \le \frac{VarY}{\alpha^2}$

Thus

$P(|X-\mathbb{E}X| > \alpha) = P(|Y| > \alpha) \le \frac{VarY}{\alpha^2} = \frac{Var X}{\alpha^2}$

>This inequality has real consequences. For example, if we pick Alpha squared equal to twice of variance x, then we see that the probability for random variable to deviate from its expected value for square root of its double variance is less than one-half. 
>We will use similar arguments in the proof of law of large numbers.

eg. let $\alpha = VarX$

$P(|X-\mathbb{E}X| > \sigma) \le \frac{1}{2}$
