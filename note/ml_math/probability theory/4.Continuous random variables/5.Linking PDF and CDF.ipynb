Relation between PDF and CDF

$CDF_X(x) = F(x)$

$PDF_X(x) = p(x)$

$P(X \in [a, b]) = F(b) - F(a)$

$p(x) = \lim_{\Delta x \to 0} \frac{P(X \in [x, x+ \Delta x])}{\Delta x} = 
\lim_{\Delta x \to 0} \frac{F(x + \Delta x) - F(x)}{\Delta x}
= \lim_{\Delta x \to 0} \frac{\Delta F}{\Delta x}= F^{\prime}(x)$

PDF is a derivative of CDF  
CDF is antiderivative of PDF

$\int^b_a p(x)dx = F^{\prime} dx = F(b)-F(a)$

_fundamental theory of calculus_



__example__

1\. unifrom distribution [a, b]

$\int^{+\infty}_{-\infty}p(x)dx=1$

$\int^{b}_{a}p_{\circ}dx=1$

$(b-a)p_{\circ} = 1 \implies p_{\circ} = \frac{1}{b-a}$

$p(x) = 
\begin{cases}
0, x \notin [a,b] \\
\frac{1}{b-a}, x\in[a,b]
\end{cases}
$

2\. Some non-uniform distribution on [0, 1]

A linear distribution, $(x=0, y=p_{\circ})$ and $(x=1, y=0)$

$\int^{1}_{0}p(x)dx = 1$

this integral is an area of the triangle, $1\cdot p_{\circ} \cdot \frac{1}{2}$

$1\cdot p_{\circ} \cdot \frac{1}{2} = 1 \implies p_{\circ} = 2$

3\. Normal (Gaussian) distribution

$pdf(x) = \frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}}$

$PDF_X$ is given by function

- 0 for $x < -2$
- $C(4-x^2)$ for $x \in [-2, 2]$
- 0 for $x > 2$

Find C. Enter the exact value below

$\int^{2}_{-2} C(4-x^2) = 1$

$\int^{2}_{-2} C(4-x^2) = \int^{2}_{-2}4C - \int^{2}_{-2}Cx^2
=4Cx|^{2}_{-2} - \frac{1}{3}Cx^{3} |^{2}_{-2}
= 16C - \frac{1}{3}C(8-(-8)) = 16C - \frac{16}{3}C = \frac{32}{3}C = 1
$

$C = \frac{3}{32}$

$CDF_X = C(4x - \frac{x^3}{3})+constant$

$CDF_X(2) = C(8 - \frac{8}{3})+constant = \frac{3}{32} \times \frac{16}{3} + constant
= \frac{1}{2} + constant = 1 \implies constant = \frac{1}{2}$

$CDF_X(-2) = C(-8 + \frac{8}{3})+constant = \frac{3}{32} \times -frac{16}{3} + constant
= -\frac{1}{2} + constant = 0 \implies constant = \frac{1}{2}$

$CDF_X(1) = C(4 - \frac{1}{3}) + \frac{1}{2} = C(\frac{11}{3}) + \frac{1}{2} = \frac{11}{32} + \frac{1}{2} = \frac{27}{32}$
