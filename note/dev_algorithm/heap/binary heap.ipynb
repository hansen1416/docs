__Binary minimum__

parent =< children

it stored in array in such a fashion

| parent | current | children |
|--|--|--|
| (i-1)/2 | i | 2i+1, 2i+2 |


- Add: append to the tail, and rebalance our heap, so the relation stays the same

- Delete: only from top, swap the value at the top with the value at the end. and delete from the end, and restrcuture the heap, so the relation stays the same

How to update our heap, two helper functions. sift_up and sift_down

sift_up,  

compare element with parent    
swap with parent if needed   
call from parent   

sift_down  

compare element with child      
swap with child if needed     
call from child     

when delete call sift_down    
when add call sift_up   

The time complexity of sift_uo and sift_down is the height of the binary tree

__Binary heap sort__

- Build the heap from the array  
- At each step, current minimum is at the top of the heap  
- Remove the top from the heap until the heap is empty, and append them to the new array  

If the parent is at position i, it's left and right children would be at positions 2 times i plus 1 and 2 times i plus 2 respectively. So this helps us to store our heap as an array.

heapify(arr, i, n)

- both subtrees of element i are heaps  
- combine then into one heap under element i  
- current heap size is n

```
def buildMaxHeap(arr):
    n = len(arr)
    start = n # 2-1 # targeting vertex would be n times 2 minus 1 is the first vertex where it shouldn't have any children.
    
    for i in range(start, -1, -1):
        heapify(arr, i, n)
```

```
def heapify(arr, i, n)
    largest = i
    l, r = i * 2 + 1, i * 2 + 2
    
    for child in (l, r):
        if child < n and arr[largest] < arr[child]:
            largest = child
            
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, largest, n)
    
```

__In-place solution:__

- Build a max heap from input array in-place (!)   
- Until the size of the heap is exactly 1 do:
    - swap the top element of heap (current maximum) and the last element of the heap  
    - decrease the size of heap by 1 (by pop from the end of the array, which is the maximum from the step1)  
    - restore the heap order: call the heapify procedure from the root
    
---
```
def heapSort(arr):
    n=len(arr)
    buildMaxHeap(arr)
    for i in range(n-1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        heapify(arr, 0, i)
```

---

__Time complexity__: O(n log n)  
__Space complexity__: O(1)
