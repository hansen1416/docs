__Binary minimum__

parent =< children

it stored in array in such a fashion

| parent | current | children |
|--|--|--|
| (i-1)/2 | i | 2i+1, 2i+2 |


- Add: append to the tail, and rebalance our heap, so the relation stays the same

- Delete: only from top, swap the value at the top with the value at the end. and delete from the end, and restrcuture the heap, so the relation stays the same

---

How to update our heap, two helper functions. sift_up and sift_down

sift_up,  

compare element with parent    
swap with parent if needed   
call from parent   

sift_down  

compare element with child      
swap with child if needed     
call from child     

when delete call sift_down    
when add call sift_up   

The time complexity of sift_uo and sift_down is the height of the binary tree

---

Well, you can just start with an empty one and then add elements one after another. But well, that's another best way to do it, you can actually just put your elements in the integer array without worrying about the relationship between the parents and the children and then you can call sift-up from each one of them. So first from the first ones, from the second, from the third, and so on. You can construct your heap in such a fashion until it will actually be more efficient than to construct it using just adding elements to your empty heaped one after another.

__Binary heap sort__

- Build the heap from the array  
- At each step, current minimum is at the top of the heap  
- Remove the top from the heap until the heap is empty, and append them to the new array  

If the parent is at position i, it's left and right children would be at positions 2 times i plus 1 and 2 times i plus 2 respectively. So this helps us to store our heap as an array.

heapify(arr, i, n)

- both subtrees of element i are heaps  
- combine them into one heap under element i  
- current heap size is n

```
def buildMaxHeap(arr):
    n = len(arr)
    start = n # 2-1 # targeting vertex would be n times 2 minus 1 is the first vertex where it shouldn't have any children.
    
    for i in range(start, -1, -1):
        heapify(arr, i, n)
```

```
def heapify(arr, i, n)
    largest = i
    l, r = i * 2 + 1, i * 2 + 2
    
    for child in (l, r):
        if child < n and arr[largest] < arr[child]:
            largest = child
            
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, largest, n)
    
```

__In-place solution:__

- Build a max heap from input array in-place (!)   
- Until the size of the heap is exactly 1 do:
    - swap the top element of heap (current maximum) and the last element of the heap  
    - decrease the size of heap by 1 (by pop from the end of the array, which is the maximum from the step1)  
    - restore the heap order: call the heapify procedure from the root
    
---
```
def heapSort(arr):
    n=len(arr)
    buildMaxHeap(arr)
    for i in range(n-1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        heapify(arr, 0, i)
```

__Time complexity__: O(n log n)  
__Space complexity__: O(1)

````python
class MaxHeap:
    def __init__(self):
        self._heap = []
        
    def push(self, elem):
        self._heap.append(elem)
        self._sift_up(len(self._heap) - 1)
    
    def _sift_up(self, index):
        # YOUR CODE GOES HERE
        # compare element with parent
        # swap with parent if needed
        # call from parent
        if index <= 0:
            return

        parentIndex = (index-1)//2
        childIndex = index

        parent = self._heap[parentIndex]
        child = self._heap[childIndex]

        # print(parentIndex, childIndex, parent, child)
        # print('before', parentIndex, childIndex, self._heap)

        if child > parent:
            self._heap[parentIndex], self._heap[childIndex] = self._heap[childIndex], self._heap[parentIndex]
            # print('after', self._heap)
            self._sift_up(parentIndex)


heap = MaxHeap()
heap.push(1)
heap.push(2)
heap.push(3)

print(heap._heap)

# check that your code works correctly on provided example
assert heap._heap in [[3, 2, 1], [3, 1, 2]], 'Wrong answer'
  
````

````python
class MinHeap:
    def __init__(self):
        self._heap = []
        
    def pop(self):
        self._heap[0], self._heap[-1] = self._heap[-1], self._heap[0]
        elem = self._heap.pop()
        self._sift_down(0)
    
    def _sift_down(self, index):
        # YOUR CODE GOES HERE
        # compare element with child
        # swap with child if needed
        # call from child
        parentIndex = index
        leftChildIndex = 2*index+1
        rightChildIndex = 2*index+2

        parent = self._heap[parentIndex]
        
        if len(self._heap) < leftChildIndex + 1:
            return
        # only left child
        elif len(self._heap) < rightChildIndex + 1:

            leftChild = self._heap[leftChildIndex]

            if leftChild < parent:
                self._heap[parentIndex], self._heap[leftChildIndex] = self._heap[leftChildIndex], self._heap[parentIndex]
                self._sift_down(leftChildIndex)
        else:
            
            leftChild = self._heap[leftChildIndex]
            rightChild = self._heap[rightChildIndex]

            if leftChild < parent and rightChild < parent:
                if leftChild < rightChild:
                    self._heap[parentIndex], self._heap[leftChildIndex] = self._heap[leftChildIndex], self._heap[parentIndex]
                    self._sift_down(leftChildIndex)
                else:
                    self._heap[parentIndex], self._heap[rightChildIndex] = self._heap[rightChildIndex], self._heap[parentIndex]
                    self._sift_down(rightChildIndex)

            elif leftChild < parent:
                self._heap[parentIndex], self._heap[leftChildIndex] = self._heap[leftChildIndex], self._heap[parentIndex]
                self._sift_down(leftChildIndex)
            elif rightChild < parent:
                self._heap[parentIndex], self._heap[rightChildIndex] = self._heap[rightChildIndex], self._heap[parentIndex]
                self._sift_down(rightChildIndex)


heap = MinHeap()
heap._heap = [1, 2, 3,4,5,6,7,8]
heap.pop()

print(heap._heap)

# check that your code works correctly on provided example
# assert heap._heap == [2, 3], 'Wrong answer'
````
