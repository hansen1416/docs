```python
class Node:
    def __init__(self, val, link=None):
        self.val=val
        self.next=link
```

```python
class LinkedList:
    def __init__(self):
        self.head=None
        
    def push_front(self, val):
        node = Node(val)
        node.next = self.head
        self.head=node
        
    def pop(self, val):
        if self.head is None:
            return
        
        if self.head.val == val:
    
            self.head = self.head.next
            return
        
        node = self.head
        # we need to remember previous node, because single node
        prev_node = self.head.next
        
        while node:
            if node.val == val:
                #remove node, only need to link the previous element to next
                prev_node.next = node.next
                del node
                return
            
            prev_node, node = node, node.next
                
        
    def search(self, val):
        node = self.head
        # treverse the entire linked list
        while node:
            if node.val == val:
                return node
            node = node.next
        
        return None
    
    def tolist(self):
        arr = []
        node = self.head
        
        while node:
            arr.append(node.val)
            node = node.next
        
        return arr
```

```python
llist = LinkedList()
llist.push_front(3)
llist.push_front(2)
llist.push_front(4)
llist.push_front(0)

print(llist.tolist())

llist.pop(4)

print(llist.tolist())

print(llist.search(2).val)
```

```python
def median_linked_list(head):
    slow, fast = head, head # left and right pointer
    
    # move them both at the same time
    # `fast` is 2 times faster than 'slow'
    # at the end, slow at the half of the linked list
    while fast:
        fast = fast.next
        if fast is not None:
            fast = fast.next
            slow = slow.next
        
    return slow
```

```python
def has_cycles(head):
    slow, fast = head, head # left and right pointer
    
    # move them both at the same time
    # `fast` is 2 times faster than 'slow'
    # when `fast` catch `slow`, it means there is a cycle
    while fast:
        fast = fast.next
        if fast is not None:
            fast = fast.next
            slow = slow.next
        if fast == slow:
            return True
            
    return False
```
