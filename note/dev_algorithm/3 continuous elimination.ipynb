```python
def lines(a):
    """
    In lines computer game, a player sets balls of different colors in a line. 
    When a continuous block of three or more balls of the same color is formed, it is removed from the line. 
    In this case, all the balls are shifted to each other, and the situation may be repeated.
    
    There can be at most one continuous block of three or more same-colored balls at the initial moment.
    
    The function takes a list a with initial balls disposition. 
    Balls number is less or equals 1000, balls colors can be from 0 to 9, each color has its own integer.
    
    The function has to return one number, the number of the balls that will be destroyed.
    """
    color = None
    i = 0
    left = 0
    right = 0
    result = 0
    temp = 0

    list_length = len(a)

    while i < list_length - 2:
        if a[i] == a[i+1] and a[i] == a[i+2]:
            color = a[i]
            left = i-1
            right = i+3
            result += 3
            break
        i+=1
    # if didn't find any continuous block initially, return 0
    if result == 0:
        return result
            
    # print(color, left, right)
    while left >= 0 or right < list_length:
        # if left/right has the same color as the squence, move one step left/right
        if left >= 0 and a[left] == color:
            # print('one step left', left, color, a[left])
            result += 1
            left -= 1
            continue
        if right < list_length and a[right] == color:
            # print('one step right', color)
            result += 1
            right += 1
            continue

        # if left and right doesn't have the same color, 
        if left >=0 and right < list_length and a[left] == a[right]:
            # print('left: {} and right: {} equal'.format(left,right), color, a[left-1], a[left], a[left-2])
            if left > 0 and a[left-1] == a[left]:
                # print('go left')
                color = a[left]
                result += 3
                left -= 2
                right += 1
                continue
            if right < list_length - 1 and a[right] == a[right + 1]:
                # print('go right')
                color = a[right]
                result += 3
                left -= 1
                right += 2
                continue

        break
        
    return result
        


# some test code
if __name__ == "__main__":
    test_a = [2,2,1,3,2,0,1,1,1,1,1,0,0,2,2,3,3,1,1,2,3,2,2,1,1,3,4,5,6]
    # should print 20
    print(lines(test_a))

    test_a = [1,0,3,0,4,0,0,5,0,1,1,2,1,1,1,1,1,1,1,2,2]
    # should print 10
    print(lines(test_a))

    test_a = [2,2,2,2,2,2,2,1,0,3,0,4,0,0,5,0,1,1,2,1,1]
    # should print 7
    print(lines(test_a))

    test_a = [1,3,3,2,2,2,3,3,1,1,4]
    # should print 10
    print(lines(test_a))

    test_a = [1,2,2,1, 2,8,3,3,9, 3, 1, 4,9,9,9]
    # should print 3
    print(lines(test_a))

    test_a = [2, 1,1,7,7,7,1,2,2,1, 2,8,3,3,9, 3, 1, 4,9,9,8,9]
    # should print 9
    print(lines(test_a))

    test_a = [2,1,2,2,1,1,8,1] * 20
    # should print 3
    print(lines(test_a))

    test_a = [2, 2,1,1,1,3,2,1,2,2]
    # should print 3
    print(lines(test_a))
```
