```python
import sys, time

classes = ('Paladin', 'Sorceress', 'Barbarian')

def create_hero(**kwargs):
    """
    assign properties to hero, eg. name, class
    """
    hero = {}
    for key, value in kwargs.items():
        hero[key] = value

    return hero

def move(hero, direction):
    """
    press w to move forward, and s to move backward
    """
    if direction == 'forward':
        hero['position'] = hero['position'] + 1
    elif direction == 'backward':
        hero['position'] = hero['position'] - 1
        if hero['position'] < 0:
            hero['position'] = 0

def inventory(hero, **items):
    """
    add invetory
    """
    for name, item in items.items():
        hero['inventory'][name] = item

    return hero['inventory']

def combination(hero, combinations):
    """
    possible combinations
    """
    if hero['clas'] == classes[0]:
        if combinations == ('q', 'w', 'e'):
            casting('TTTTTTTTTTTTTTTT')
            return True
        elif combinations == ('a', 's', 'd'):
            casting('LLLLLLLLLLLLLLLL')
            return True
    elif hero['clas'] == classes[1]:
        if combinations == ('q', 'w', 'e'):
            casting('--------------------')
            return True
        elif combinations == ('a', 's', 'd'):
            casting('-*-*-*-*-*-*-*-*-*-*')
            return True
    elif hero['clas'] == classes[2]:
        if combinations == ('q', 'w', 'e'):
            casting('/\/\/\/\/\/\/\/\/\/\\')
            return True
        elif combinations == ('a', 's', 'd'):
            casting('____IIII____IIII')
            return True
    
    return False

def casting(spell):
    if mode == 'd':
        spell = spell[0:4]
    for s in spell:
        sys.stdout.write(s)
        sys.stdout.flush()
        time.sleep(0.05)

mode=input('Choose a mode e: easy (default), d: difficult, ')

if mode == 'd':
    print("Starting with mode Difficult.")
else:
    print("Starting with mode Easy.")

name=input("Type a name for your hero, ")
clas_sort=input("Choose a class for you hero, p: "+ classes[0] +" s: " + classes[1] + ", b: " + classes[2] + "  (default), ")

if clas_sort == 'p':
    clas = classes[0]
elif clas_sort == 's':
    clas = classes[1]
else:
    clas = classes[2]

Hero = create_hero(name = name, clas=clas, position = 0,  inventory = {})

print("Your " + Hero['clas'] + ' ' + Hero['name'] + ' is ready.')
print('Adventure start.. press \'w\' move forward, \'s\' move backward')

direction=input()

if (direction == 'w'):
    move(Hero, 'forward')
elif (direction == 's'):
    move(Hero, 'backward')

pick = input('Pick up new items. y/n ')

if pick == 'y':
    if mode == 'd':
        inventory(Hero, weapon = 'Wood sword', shield = 'Wood shield')
    else:
        inventory(Hero, weapon = 'Dragon sword', shield = 'Tower shield')

print("invenory", inventory(Hero))

print('       =。=     An enemy emerged, start casting!! (q,w,e) (a,s,d)')

comb = (input(), input(), input())

result = combination(Hero, comb)

if result:
    print('\nEmemy eliminated!')
    print('GOOD GAME!')
else:
    print('\nCasting failed! Lost one life')
    print('GAME OVER')
```
