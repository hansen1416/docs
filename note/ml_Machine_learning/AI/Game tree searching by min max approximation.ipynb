## 2. Generalized Mean Values

$ let\:a = (a_1, a_2, ... a_n) $ be a vector of positive number  
let p be nonzero real number   
- generalized p-mean of a, $ M_p(a) $  
$ M_p(a) =  \Big(\frac{1}{n}\sum_{i+1}^{n} a_i^p\Big) $  
$ M_1(a) $ is the orinary arithmetic mean


let p = 0

- $ M_0(a) = \lim_{p=0} M_p(a) = (a_1...a_n)^{\frac{1}{n}} $  
$ M_0(a) $ is the geomatric mean


begin wtih the fact:  
$ p < q \implies M_p(a) \leq M_q(a) $ where equality only holds on the right if all $a_i$ are equal.

we are most intersted in the following two facts

$ \lim_{p \to \infty} \: M_p(a) = max(a_1,...,a_n) $   
$ \lim_{p \to -\infty} \: M_p(a) = min(a_1,...,a_n) $

For large positive or negative value of *p*, $M_p(a)$ is a good approximation to $max_i(a_i) \:or\: min_i(a_i)$, respectively, However, M_p(a), unlike max or min, has continuous derivatives with respect to each variable $a_i$.
The partial derivative of $M_p(a)$ with respect to a is:  

$ \frac{\partial M_p(a)}{\partial a_i} \; = \; \frac{1}{n}\Big(\frac{a_i}{M_p(a)}\Big)^{p-1} $  

The major reason that the generlized means are of interest to us here is that they are more suitable for "sensitivity analysis" than the min or max functions. We propose that $\partial M_p(a) / \partial a_i$ for large *p* is a more useful quantity than $\partial max(a) / \partial a_i$ since the latter is zero unless $a_i$ is maximum, in which case it is one. This discontinuous  behavior is awkward to work with, where as the partial derivative is continuous. We also note that the partial derivative ranges in value from 0 to $n^{-1/p} \; \approx \; 1 \; for \; p \gg ln \; n $

### 3.2 searching a game tree

- when C is small, the tree can be explored completely, so optimal play is possible .
- On slightly larger tree, with apha-beta pruning, may produce optimal play even though only a small fraction of tree is searched.
- However, for most interesting games, the game tree need heuristic approximations.
- A heuristic method is usually based on a "static evaluation function" $\hat{v}$ that gives a estimate $\hat{v}(a)$ of the backed-up value v(c) for a nonterminal node c. The estimate is based on "static" feature of the current  configuration that can be evaluation without further look-ahead.
- *$ B^* algorithm $* requires two static evaluation functions which are upper and lower bounds on $v(\cdot)$
- *iterative deepening* for a very large game tree, select a suitable depth bound *d*, to estimate the values of the nodes at the depth *d* using the static evaluator, and the calculate the baked-up minmax value from the nodes at depth *d* using alpha-beta pruning  
    Given a limit on the computing time available, one can successively compute the backed-up values for depth *d* = 1,2,..., until one runs out of time, using the move determined by the last search completed.
