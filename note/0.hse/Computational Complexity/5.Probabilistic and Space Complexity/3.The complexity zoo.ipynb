## The class coNP

### Chip design

Two circuts  
Do they compute the same Boolean function?

- A useful question when optimizing the microprocessors.
    Do the optimized component (size reduced) still compute the same Boolean function?
    
- This problem is not known to be in NP
    Thus, not known how to "prove equivalence efficiently"
    
- However the complement of the language is in NP:
    Certificate = an input where values differ.
    
- We have "No-witnesses".

Definition:

__A problem is in coNP if its complement is in NP__

eg. A language $L$ over $\sigma$ is in coNP if $\Sigma^* \ L$ is in NP.

__coNP-completeness__

Definition: A problem is coNP-complete if it is in coNP and every problem in coNP reduces to it.

Theorem: A problem is NP-complete if and only if its complement is NP complete.

Lemma: $L$ is NP-complete $\implies$ $L^c$ is coNP-complete

_Proof_. If $A \le_P L$ then $A^c \le_P L_c$

If A reduces to L, then the complement of A also reduces to the complement of L. This is just the same requirement, you can compute the reduction polynomial time and X is in A, if and only if (F of X is in L)?, but this is true for complements as well.

Circuts $\phi$ and $\psi$     
Is $\phi(x) = \psi(x)$ for all inputs?

Corollary: Circut-Equivalence is coNp-complete.

_Proof_

- The problem is in coNP. Indeed, the complement asks whether $\phi \ne \psi$  
    Certificate = an input x with $\phi(x) \ne \psi(x)$

- Hardness. Circut-unsatisfiability problem is coNP-complete (because Circut-satisfiability problem is NP-complete)
    Circut-unsatisfiability is a special case of Circut-Equivalence: Set $\psi=0$.  
    $\forall x[\phi(x)=0] \iff \forall x[\phi(x)=\psi(x)]$
    
(so similar as NP, to prove coNp-complete, one must prove coNP amd coNP-hard)

## NP-intermediate classes

$NP \cap coNP$

(problems outside of this intersection is NP-complete and coNP-complete)

e.g.

__The small-factor problem__

Given 2 intergers a and b  
Does a have a positive divisor smaller then b?

If there is a poly-time solution for this problem.

- The given an integer, one can compute all its factors in poly time. (By binary search, we can compute all the divisors of b. We can compute a prime factorization of a.)

- One can break ctyptography

__What if small-factor is NP-complete?__

Becasue  small-factor $\in$ coNP.

Theorem: If a problem in coNp is NP-complete, then coNP=NP.

_Proof_.

If $A \le_P B$ and $B \in coNP$, then $A \in coNP$

Hence, all problems in NP reduces to coNP, if B is NP-complete. So $NP \subset coNP$.  
Similarly, we can prove $coNP \subset NP$. 

Thus if small-factor is NP-complete, then NP=coNP.

__NP-intermediate__

Definition: A problem is NP-intermediate if it is neither in P and neither in NP-complete.

- Conjecture: small-factor problem is NP-intermediate.

- Proving that a problem is NP-intermediate implies that P$\ne$NP

## The class coRP

Definition: A problem L is in coRP if it is the complement of a problem in RP. (Randomized polynomial time)

There exists a poly time algorithm that on input x
- with $x \in L$ outputs True with prob. 1
- with $x \notin L$ outputs False with prob. 2/3

coRP $\subseteq$ coNP

__example: equivalence of polynomials__

Given 2 polynomials in serverl variables written with additions and multiplications (no powers).  
Are they equal for all interger inputs?

Input polynomials p and q over variables $x_1, \dots, x_m$
- d the maximal degree of p,q
- select $x_1, \dots, x_m$ randomly in $\{1,\dots,3d\}$
- evaludate the polynomials.
- if the value differ, output False (reject), otherwise True (accept).

Runs in polynomial time, becasue d and m are at most polynomial. (Each multiplication increases the bit size of the number by at most a logarithmic number of bits, so $\log n$ bits. In total, the largest value we can end up is something has bit size about $d \log n$)  

if p(x)=q(x) for all x, then it always output True.

Need to prove: $p(x) \ne q(x)$, the output is False with prob. $\ge 2/3$.

We do it only for special case p and q with a single variable.

- The polynomial $p(x)-q(x)$ is nonzero and has degree $\le d$
- Thus it is 0 for at most d defferent x
- Hence, with prob. $\le 1/3$, a random $x \in \{1,\dots,3d\}$ satisfies $p(x)-q(x)=0$.
- The outputs is True with probability $\le 1/3$



Definition: A language is in ZPP if it can be decided by a Las Vegas algorithm with expected running time at most a polynomial in the input length.

Theorem: $ZPP = coRP \cap RP$

## P $\ne$ EXP

(polynomial is not equal exponential)

Definition: (exponential time decidable languages)

$EXP=\bigcup_i TIME(2^{n^i})$

clearly $P \subset EXP$. the upper bound is much larger.

__proof $\exists D \in EXP\backslash P$__

List of all turing machines $M_1, M_2, \dots$

In a typical enumeration, equivalent copies of a machine appear infinitely often.

Simulating machine $M_i$ on input $w$ for $2^{|w|}$ steps takes time 

$poly(i, |w|) \cdot 2^{|w|} \le 2^{O(|w|+\log i)}$

Let $w_i$ be the $i$th string in $\varepsilon$, 0,1,00,01,10,11,...  
Note $\log i \le |w_i|$. For input $w_i$, this is bounded by 

$2^{O(|w_i|)}$

Inverse diagonal language

$w_i \notin D \iff M_i$ accepts $w_i$ in $\le 2^{|w_i|}$ steps.

Note $D\in EXP$. If runtime $(M_i(w_i)) \le 2^{|w_i|}$, then

$w_i \notin D \iff M_i$ accepts $w_i$

_Proof_ thet if $L \in D$ then $L \notin D$.

Let $\tilde{M}$ be a decider of $L$ running in time $p(n)$ where $p$ is a polynomial. Let $M_a,M_b,M_c,...$ be equivalent copies of $\tilde{M}$.

For all large $i \in \{a,b,c,...\}$ we have

$p(|w_i|) \le 2^{|w_i|}$

Thus for such $i$, runtime$(|w_i|)\le 2^{|w_i|}$, and the equivalence is satisfied.

__Theorem: There exists a language in EXP that is not in P__

Similar for space comlexity classes

EXSPACE = $\bigcup_cSPACE(2^{n^c})$

With the same proof of idea, one can show that

__Theorem: There exists a language in EXSPACE that is not in PSPACE__

## Overview of complexity classes

$P \subseteq NP \subseteq PSPACE \subseteq EXP$

if $P=NP$, solve many problem in poly-time. (TSP)

if $NP = PSPACE$, many game has winning stratergy.

if $PSPACE=EXP$, very long computations, we can do in small space.

We have proven $P \subset EXP$, so one of these 3 inclusion must be strict.



$P \subseteq ZPP \subseteq RP \subseteq NP \subseteq PSPACE \subseteq EXP$
