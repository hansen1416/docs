## Blending and stacking

Ensembling is a way to combine weaker base models into the stronger one.

We used decision trees to obtain a boosting model. What if we make an ensemble out of boosting and other trained models?

What if we combine the predictions of all trained models with some higher-level model?



We cannot, train this matter algorithm on the same observations where we trained these first level models. This is why we need a Validation dataset.

Problems with blending: we need to split our data into train and validation. Use train data to train base model, validation model to train meta-algorithm.

This way, neither base models nor meta-algorithm use the whole data for training

__Stacking__

Similar as cross-validation, train one base model on some folds and predict on the other folds. And train a different model on different folds and predict on remaining folds. and so on.

At the end we have base model prediction on all training data. We can use these data to train meta-algorithm.

Stacking might improve the score of single algorithms because it is a clever way to combine machine learning models which is aimed at taking advantage of their strengths

- Need __a lot__ of data

- Need __uncorrelated base models__ - even better if powerful

- Meta-algorithm: regularized linear model, boosting (this is a relatively simpler task, less feature. Choose any algorithm that fit)

- Modifications:
    - make different stackings with different number of folds and average them
    - multi-level stacking
    - combine features with meta-algorithm


__Summary__

- Blending and stacking are powerful techniques of ensembling trained models

- If base modesl are good and uncorrelated, they may significantly improve overall results

- However, they need a lot of trainning data and less effficient than simple 1-level models

- In blending, we split the data into train and validation parts in order to be able to train a 2nd level model. In stacking, we use a cross-validation approach to be able to train a 2nd level model on the whole data

- Along with 1st level models predictions, it is possible to use regular features from the data as inputs to the 2nd level model

Stacking has potential to solve complex problems.
