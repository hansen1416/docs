Gini impurity of the node = prob left leaf x gini left + prob right leaf x gini right

__Calcuate Gini for all nodes, take the one with smallest Gini as root node__

Fix $\mathcal y$,

### Bound on the number of trees in $\mathcal H_k$ and count all descriptions

Plan: fix a method to describe trees in $\mathcal H_k$ and count all descriptions 

We bound the number of decision trees by the number of descriptions. 

Every decision tree must have at least one description, and each description can have at most one decision trees.
Some descriptions might be invalid, so there is no decision tree. 

Two procedures, from root node, go through all nodes depth first, (width first also work). write down each node, we have a list.

We can also build a tree from a node list.

One to make trees to list and one to make list to trees
