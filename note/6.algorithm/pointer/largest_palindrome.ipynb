```python
def largest_palindrome(s):

    string_length = len(s)
    i = 0
    # pointer goes to left
    left = 0
    # pointer goes to right
    right = 0

    palindrome = s[0]

    while i < string_length - 1:
        # try to find the center of palindrome
        if s[i] == s[i + 1]:
            # consider the case "DDDDDD" vs "DDDDDDD"
            if i > 0 and s[i-1] == s[i+1]:
                left = i - 1
                right = i+1
            else:
                left = i
                right = i+1
        elif i > 0 and s[i-1] == s[i+1]:
            left = i-1
            right = i+1
        else:
            # reset left and right pointer
            left = 0
            right = 0

        if right - left + 1 > len(palindrome):
            palindrome = s[left:right+1]

        left -= 1
        right += 1

        # print(left, right, len(palindrome))
        #ã€€start from the palindrome center, expand to left and right
        while right > 0 and left >= 0 and right < string_length:
            if s[left] == s[right]:
                if right - left + 1 > len(palindrome):
                    palindrome = s[left:right+1]
                left -= 1
                right += 1
            else:
                # print('mm', len(palindrome))
                break

        i += 1

    return palindrome


# some test code
if __name__ == "__main__":
    test_s = 'ABBCBBA'
    # should print ABBCBBA
    print(largest_palindrome(test_s))

    test_s = 'ABACABAD'
    # should print ABACABA
    print(largest_palindrome(test_s))

    test_s = 'ACDBVDAECADSFGG'
    # should print GG
    print(largest_palindrome(test_s))

    test_s = 'AAAAAABBAAAAAAAEEEEEEAAAAAAAAAAAAACCCCCCCDDCCCCCCCAAA'
    # should print A
    print(largest_palindrome(test_s))

    test_s = 'AACDBVDAECADSF'
    # should print AA
    print(largest_palindrome(test_s))
    
    test_s = 'DDDDDDDAVDDDDDDD'
    # should print DDDDDDD
    print(largest_palindrome(test_s))
    
```
