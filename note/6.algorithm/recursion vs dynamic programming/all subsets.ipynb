```python
def subsets(a):
    """
    return all the possible subsets of a distinct list
    """
    if len(a) == 0:
        return [[]]

    result = subsets(a[:-1])

    for i in range(len(result)):
        result.append(result[i] + [a[-1]])
    return result

print(subsets([1,2,3]))
print(subsets([1,2,3,4]))
print(len(subsets([1, 3, 5, 7, 9, 17])))
```

```python
def subsets_fix_length(elems, k):
    if k == 0:
        return [ [] ]
    if len(elems) == 0:
        return []
    last = elems[-1]
    without_last = elems[:-1]
    result = subsets_fix_length(without_last, k)
    with_last = subsets_fix_length(without_last, k - 1)
    for s in with_last:
        result.append(s + [last])
    return result
# warning, don't go over range(100)
r = subsets_fix_length(list(range(10)), 4)
print(r)
```

```python
def submessages(s):
    if len(s) == 0:
        return {''}
    last = s[-1]
    without_last = s[:-1]
#     print(s)
    result = submessages(without_last)
    for m in list(result):
        result.add(m + last)
    return result
  
# should print {"", "a", "b", "aa", "ab", "ba", "aba"}
# elements of the set can be printed in a different order
# print(submessages("ab"))

# print(submessages("abc"))
print(submessages("aba"))
```
